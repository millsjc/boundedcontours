[Mean integrated square error](https://en.wikipedia.org/wiki/Mean_integrated_squared_error) can be used to compare the quality of a kernel density estimate to the true density. Here we show that for the bounded distributions tested, a bounded estimate is better than an unbounded estimate.

```python
from boundedcontours.demo import mean_integrated_squared_error
```

```python
try:
    mean_integrated_squared_error.check_bounded_is_better(
        plot_example=False,
        plotly=False,
        sample_size=int(1e5),
        nbins=30,
        n_simulations=10,
    )
    print("MISE was lower for bounded contours in all cases.")
except ValueError as e:
    raise ValueError(
        "Warning: MISE was higher for bounded contours in some cases."
    ) from e
```
